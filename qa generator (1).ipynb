```python
import PyPDF2

def extract_text_from_pdf(pdf_file):
    with open(pdf_file, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text()
    return text

pdf_file = "/content/Data_Analytics Client Brief.pdf"
pdf_text = extract_text_from_pdf(pdf_file)
print(pdf_text[:1000])
```

```python
from transformers import pipeline
```

```python
from transformers import pipeline

# Initialize the summarization pipeline
summarizer = pipeline('summarization', model="facebook/bart-large-cnn")

def summarize_text(text):
    # Break the text into chunks if it's too long for the model
    max_chunk_size = 1024  # Max token length for the model
    text_chunks = [text[i:i+max_chunk_size] for i in range(0, len(text), max_chunk_size)]

    summary = ''
    for chunk in text_chunks:
        summarized_chunk = summarizer(chunk, max_length=150, min_length=30, do_sample=False)
        summary += summarized_chunk[0]['summary_text'] + " "
    return summary

# Summarize the extracted text from the PDF
pdf_summary = summarize_text(pdf_text)
print(pdf_summary)
```

```python
def extract_text_from_pdf_with_plumber(pdf_file): # Use the pdf_file variable as the input
    text = ''
    with pdfplumber.open(pdf_file) as pdf: # Open the file specified by the pdf_file variable
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Use pdfplumber to extract text from the PDF
pdf_text = extract_text_from_pdf_with_plumber(pdf_file) # Pass the pdf_file variable containing the correct file name
print(pdf_text[:1000])
```

```python
import random
```

```python
from transformers import pipeline
import random
qg_pipeline = pipeline("text2text-generation", model="valhalla/t5-small-qg-hl")

summary = pdf_summary
questions = qg_pipeline(summary)

question_texts = [q['generated_text'] for q in questions]

num_questions_to_print = min(10, len(question_texts))
random_questions = random.sample(question_texts, num_questions_to_print)

print("Generated Questions:")
for question in random_questions:
    print(question)
```

```python
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
for question in questions:
    answer = qa_pipeline(question=question['generated_text'], context=summary)
    print(f"Q: {question['generated_text']}\nA: {answer['answer']}")
```
